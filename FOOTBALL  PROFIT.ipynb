```python
dict_countries = {
              'Spanish La Liga':'SP1', 'Spanish Segunda Division':'SP2',
              'German Bundesliga':'D1', 
              'English Premier League':'E0', 'English League 1':'E2', 'English League 2':'E3',
             }
```

```python
dict_historical_data = {} 

# looping through key elements
for league in dict_countries:
    frames = [] # one dataframe per league
    for season in range(15, 21):
        df = pd.read_csv(root + str(season) + str(season+1) + "/" + dict_countries[league]+ ".csv")
        df.insert(1, 'season', season)
        frames.append(df)
    df_frames = pd.concat(frames) # storing all seasons of a league in a single dataframe
    dict_historical_data[league] = df_frames
```

# PROFITABLE DECISION

```python
import pandas as pd
import numpy as np
from scipy.stats import poisson

df = pd.read_csv('https://www.football-data.co.uk/mmz4281/2526/D1.csv')

avg_home_goals = df['FTHG'].mean()
avg_away_goals = df['FTAG'].mean()

teams = sorted(df['HomeTeam'].unique())
team_stats = []

for team in teams:
    home_games = df[df['HomeTeam'] == team]
    home_att = home_games['FTHG'].mean() / avg_home_goals
    home_def = home_games['FTAG'].mean() / avg_away_goals
    away_games = df[df['AwayTeam'] == team]
    away_att = away_games['FTAG'].mean() / avg_away_goals
    away_def = away_games['FTHG'].mean() / avg_home_goals
    
    team_stats.append({
        'Team': team,
        'HomeAttack': home_att,
        'HomeDefense': home_def,
        'AwayAttack': away_att,
        'AwayDefense': away_def,
        'OverallStrength': (home_att + away_att) - (home_def + away_def)
    })

stats_df = pd.DataFrame(team_stats).sort_values(by='OverallStrength', ascending=False)
stats_df.to_csv('team_strengths.csv', index=False)

def get_probs(home_team, away_team):
    h_stats = stats_df[stats_df['Team'] == home_team].iloc[0]
    a_stats = stats_df[stats_df['Team'] == away_team].iloc[0]
    
    exp_h = h_stats['HomeAttack'] * a_stats['AwayDefense'] * avg_home_goals
    exp_a = a_stats['AwayAttack'] * h_stats['HomeDefense'] * avg_away_goals
    
    h_probs = [poisson.pmf(i, exp_h) for i in range(10)]
    a_probs = [poisson.pmf(i, exp_a) for i in range(10)]
    
    m = np.outer(h_probs, a_probs)
    
    win = np.sum(np.tril(m, -1))
    draw = np.sum(np.diag(m))
    loss = np.sum(np.triu(m, 1))
    
    return win, draw, loss

# Sample prediction for the top two teams
top_team = stats_df.iloc[0]['Team']
second_team = stats_df.iloc[1]['Team']
win, draw, loss = get_probs(top_team, second_team)

print(f"Teams in dataset: {teams}")
print(f"Top team: {top_team}")
print(f"Matchup: {top_team} vs {second_team}")
print(f"Win: {win:.2%}, Draw: {draw:.2%}, Loss: {loss:.2%}")
print(stats_df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://www.football-data.co.uk/mmz4281/2526/D1.csv')
winning_rate = df.groupby('FTR').size()
winning_rate.index = winning_rate.index.map({'H': 'Home Win', 'A': 'Away Win', 'D': 'Draw'})

plt.pie(
    winning_rate, 
    labels=winning_rate.index, 
    autopct='%.1f%%', 
    startangle=90
)

plt.title('Overall Winning Rate') 
plt.axis('equal') 
plt.show()
```

## CONCLUSION

For a profitable strategy, these percentages serve as the benchmark; real money is made by identifying matches where the bookmaker's implied probability is lower than these historical winning rates.
